```python
# Count non-null values for each row
df['NUM_VEHICLES'] = df[['VEHICLE TYPE CODE 1', 'VEHICLE TYPE CODE 2', 
                         'VEHICLE TYPE CODE 3', 'VEHICLE TYPE CODE 4']].notnull().sum(axis=1)

# Plot histogram
df['NUM_VEHICLES'].value_counts().sort_index().plot(kind='bar', figsize=(8, 5), color='orange')
plt.title('Number of Vehicles Involved per Crash')
plt.xlabel('Number of Vehicles')
plt.ylabel('Count')
plt.show()
```
