```python
df = pd.read_csv("/Users/harsnyi/Documents/NYC-Crash-Analysis/data/NYC_weather.csv")
df_crash = pd.read_csv("/Users/harsnyi/Documents/NYC-Crash-Analysis/data/NYC_crashes.csv")
```

```python
df['time'] = pd.to_datetime(df['time'], errors='coerce')
```

```python
df_crash['crash_time'] = pd.to_datetime(df_crash[['YEAR', 'MONTH', 'DAY', 'HOUR']])
```

```python
df.info()
```

```python
df = df.sort_values('time')
df_crash = df_crash.sort_values('crash_time')

# Perform an asof merge
df_combined = pd.merge_asof(
    df_crash,  # Left dataframe (crashes)
    df,  # Right dataframe (weather)
    left_on='crash_time',  # Key for left dataframe
    right_on='time',  # Key for right dataframe
    direction='backward'  # Match the closest earlier weather record
)
```
